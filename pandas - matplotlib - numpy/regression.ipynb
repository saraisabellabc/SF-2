write a snippet of code that returns the minimum value in the 'Scores' column

```python
grades_df['Scores'].min()

min = 1000000
for i in range(len(grades_df['Scores'])):
    if grades_df['Scores'].iloc[i] < min:
        min = grades_df['Scores'].iloc[i]
print(min)
```
